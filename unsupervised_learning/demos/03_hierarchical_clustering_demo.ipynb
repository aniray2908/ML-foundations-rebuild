# Hierarchical Clustering Demonstration

Goal:
- Visualize dendrogram structure
- Compare linkage methods
- Observe chaining effect
- Compare behavior on non-spherical data

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

plt.scatter(X[:,0], X[:,1])
plt.title("True Spherical Clusters")
plt.show()
```

```python
Z_complete = linkage(X, method='complete')

plt.figure(figsize=(8,5))
dendrogram(Z_complete)
plt.title("Dendrogram - Complete Linkage")
plt.show()
```

```python
Z_single = linkage(X, method='single')

plt.figure(figsize=(8,5))
dendrogram(Z_single)
plt.title("Dendrogram - Single Linkage")
plt.show()
```

```python
Z_single = linkage(X, method='single')

plt.figure(figsize=(8,5))
dendrogram(Z_single)
plt.title("Dendrogram - Single Linkage")
plt.show()
```

```python
model_complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
labels_complete = model_complete.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels_complete)
plt.title("Complete Linkage Clustering")
plt.show()
```

```python
np.random.seed(42)
bridge = np.random.uniform(low=[-2, -2], high=[6, 6], size=(30,2))
X_bridge = np.vstack([X, bridge])

plt.scatter(X_bridge[:,0], X_bridge[:,1])
plt.title("Clusters with Noise Bridge")
plt.show()
```

```python
np.random.seed(42)
bridge = np.random.uniform(low=[-2, -2], high=[6, 6], size=(30,2))
X_bridge = np.vstack([X, bridge])

plt.scatter(X_bridge[:,0], X_bridge[:,1])
plt.title("Clusters with Noise Bridge")
plt.show()
```

```python
X_moon, _ = make_moons(n_samples=300, noise=0.05, random_state=42)

plt.scatter(X_moon[:,0], X_moon[:,1])
plt.title("Moon Data")
plt.show()
```

```python
model_complete_moon = AgglomerativeClustering(n_clusters=2, linkage='complete')
labels_complete_moon = model_complete_moon.fit_predict(X_moon)

plt.scatter(X_moon[:,0], X_moon[:,1], c=labels_complete_moon)
plt.title("Complete Linkage on Moon Data")
plt.show()
```

```python
model_single_moon = AgglomerativeClustering(n_clusters=2, linkage='single')
labels_single_moon = model_single_moon.fit_predict(X_moon)

plt.scatter(X_moon[:,0], X_moon[:,1], c=labels_single_moon)
plt.title("Single Linkage on Moon Data")
plt.show()
```

Observations:

1. Dendrogram height represents merge distance.
2. Complete linkage produces compact clusters.
3. Single linkage produces chaining behavior.
4. Noise bridges strongly affect single linkage.
5. Hierarchical clustering does not assume spherical clusters.
6. Linkage choice defines cluster geometry.

Hierarchical clustering reveals structure before deciding cluster count.
